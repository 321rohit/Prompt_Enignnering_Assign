***Load the Transformer Model's Encoder***

```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")
```

***Prepare a List of Sample User Queries***

```python
queries = [
    "How is the weather today?",
    "Will it be rail,hot or cold tomorrow?",
    "How weather will be next week?",
    "Is there a chance of rain next week?",
    "How was the weather and temperature yesterday?"
]
```

```python
import torch
```

***Encode Each Query Using the Encoder Model***

```python
encoded_queries = []

for a in queries:
    inputs = tokenizer(a, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    encoded_query_result = outputs.last_hidden_state[:, 0, :].squeeze().numpy()
    encoded_queries.append(encoded_query_result)
```

***Generate Prompts for the AI to Respond To***

```python
prompts_generate = [
    "The current weather is sunny with a high of 45°C.",
    "Tomorrow will be mostly hot and sunny.",
    "It will be very hot next week, with temperatures reaching up to 48°C.",
    "Rain is unlikely next week, with temperatures expected to stay above 45°C.",
    "Yesterday, the temperature was around 42°C,sunnt and hot."
]
```

***Displaying both user queries and generated prompts***

```python
for query, encoded_query, prompt in zip(queries, encoded_queries, prompts_generate):
    print(f"User Query: {query}")
    print(f"Generated Prompt: {prompt}")
    print("-" * 50)
```
